## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
#     return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image

# def reject_outliers(data, m=1):
#     data = np.array(data)
#     d = abs(data - np.median(data))
#     stdev = np.std(data)
#     return data[ d < m * np.std(data)]

# def reject_outliers(data, m=1):
#     data = np.array(data)
#     d = abs(data - np.median(data))
#     stdev = np.std(data)
# #     print('stddev', stdev)
# #     print('d', d)
#     return data[ d < m * np.std(data)]

def draw_lane_lines_latest(img, lines, color=[255,0,0], thickness=10):
    left = []
    right = []
    ymin = img.shape[0]
    for line in lines:
        for x1,y1,x2,y2 in line:
            # get min value of y coordinate for both lines.
            if (y1 < ymin):
                ymin = y1
            if (y2 < ymin):
                ymin = y2 
            # compute slope and filter out those between -0.3 to 0.3. 
            # Add rest to left & right buffers.    
            slope = ((y2-y1)/(x2-x1))    
            if slope < -0.3:
                left.append(line)
            elif slope > 0.3:
                right.append(line)
  
    # compute mean of left line segments & right line segments.
    left_mean = np.mean(left, axis=0)
    right_mean = np.mean(right, axis=0)
    
    # compute left & right slopes using the above means.
    left_slope = (left_mean[0][3] - left_mean[0][1])/(left_mean[0][2] - left_mean[0][0])
    right_slope = (right_mean[0][3] - right_mean[0][1])/(right_mean[0][2] - right_mean[0][0])
    
    # determine start & end coordinates of the left & right line
    x1_left = (ymin - left_mean[0][1])/left_slope + left_mean[0][0]
    x2_left = (img.shape[0] - left_mean[0][3])/left_slope + left_mean[0][2]
    
    x1_right = (ymin - right_mean[0][1])/right_slope + right_mean[0][0]
    x2_right = (img.shape[0] - right_mean[0][3])/right_slope + right_mean[0][2]

    cv2.line(img, (int(x1_left), ymin), (int(x2_left), img.shape[0]), color, thickness)
    cv2.line(img, (int(x1_right), ymin), (int(x2_right), img.shape[0]), color, thickness)

def reject_outliers(data, m = 1.0):
    data = np.array(data)
    med = np.median(data)
    sd = np.std(data, axis=0)
    lt = med - (m*sd)
    gt = med + (m*sd)
    temp = np.copy(data)
    return temp[(lt <= data) & (data <= gt)]
    
def draw_lane_lines(img, lines, color=[255,0,0], thickness=2):
    left = []
    right = []
    left_slopes = []
    right_slopes = []
    ymin = img.shape[0]
    
    for line in lines:
        for x1,y1,x2,y2 in line:
            if (y1 < ymin):
                ymin = y1
            
            if (y2 < ymin):
                ymin = y2    

            slope = ((y2-y1)/(x2-x1))
            if slope < 0:
                left_slopes.append(slope)
            else:
                right_slopes.append(slope)
#     print('before - left', left_slopes)
#     print('before - right', right_slopes)
    left_slopes = reject_outliers(left_slopes)            
#     print('after-left', left_slopes)
    right_slopes = reject_outliers(right_slopes)            
#     print('after-right', right_slopes)
    l_ms = np.mean(left_slopes, axis=0)
    r_ms = np.mean(right_slopes, axis=0)
    for line in lines:
        for x1,y1,x2,y2 in line:
#             print(slope)
            slope = ((y2-y1)/(x2-x1))
            if slope < 0:
                if slope in left_slopes:
                    left.append(line)
            elif slope > 0:
                if slope in right_slopes:
#                     print('right', right_slopes)
                    right.append(line)
    
    left_mean = np.mean(left, axis=0)
    right_mean = np.mean(right, axis=0)

#     if len(left) == 0:
#         left_mean = prev_left
#     else:    
#         left_mean = np.mean(left, axis=0)
#         prev_left = left_mean
        
#     if len(right) == 0:    
#         right_mean = prev_right
#     else:    
#         right_mean = np.mean(right, axis=0)
#         prev_right = right_mean
        
#     print('left-mean', left, left_mean)
#     print('right-mean', right, right_mean)

    x1_left = (ymin - left_mean[0][1])/l_ms + left_mean[0][0]
    x2_left = (img.shape[0] - left_mean[0][3])/l_ms + left_mean[0][2]
    
    x1_right = (ymin - right_mean[0][1])/r_ms + right_mean[0][0]
    x2_right = (img.shape[0] - right_mean[0][3])/r_ms + right_mean[0][2]

    cv2.line(img, (int(x1_left), ymin), (int(x2_left), img.shape[0]), color, 6)
    cv2.line(img, (int(x1_right), ymin), (int(x2_right), img.shape[0]), color, 6)
        
def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
#     draw_lines(line_img, lines)
    draw_lane_lines_latest(line_img, lines)
#     draw_lane_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, =0.8, =1., =0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img *  + img *  + 
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, , img, , )
````

````python
def renderLane(image):
    region_select = np.copy(image)
    # plt.imshow(image)
    gray = grayscale(image)
    kernelSize = 5
    blur_gray = gaussian_blur(gray, 5)
    low_threshold = 50
    high_threshold = 150
    edges = canny(blur_gray, low_threshold, high_threshold)
#     plt.imshow(edges, cmap='Greys_r')

    imshape = image.shape
    vertices = np.array([[(50,imshape[0]),(450, 330), (510, 320), (imshape[1],imshape[0])]], dtype=np.int32)

    # original with the mask
    original_masked = region_of_interest(region_select, vertices)
#     plt.figure()
#     plt.imshow(original_masked)

    # masked-grayscale img.
    masked_img = region_of_interest(edges, vertices)
    plt.figure()
    plt.imshow(masked_img)

    rho = 2 # distance resolution in pixels of the Hough grid
    theta = np.pi/180 # angular resolution in radians of the Hough grid
    threshold = 15     # minimum number of votes (intersections in Hough grid cell)
    min_line_length = 30 #minimum number of pixels making up a line
    max_line_gap = 20    # maximum gap in pixels between connectable line segments

    lines = hough_lines(masked_img, rho, theta, threshold, min_line_length, max_line_gap)
#     color_edges = np.dstack((edges, edges, edges)) 
#     result = weighted_img(lines, color_edges)
    result = weighted_img(lines, image)
    plt.figure()
    plt.imshow(result)
````
